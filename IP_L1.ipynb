```python
from skimage import io
import numpy as np
from matplotlib import pyplot as plt
my_image = io.imread('sample1.jpg')
print(my_image)
plt.imshow(my_image)
print(my_image.shape)
print(my_image.size)
print(type(my_image))
print(my_image.min(), my_image.max())
```

```python
random_image = np.random.random([500,500])
plt.imshow(random_image)
print(random_image)
print(random_image.min(), random_image.max())
print(random_image.shape)
print(random_image.size)
print(type(random_image))
```

```python
from skimage import img_as_float
my_float_image = img_as_float(my_image)
print(my_float_image.min(), my_float_image.max())
```

```python
from skimage import img_as_int
my_int_img = img_as_int(random_image)
plt.imshow(my_int_img)
print(my_int_img)
print(my_int_img.min(), my_int_img.max())
print(my_int_img.shape)
print(my_int_img.size)
print(type(my_int_img))
```
